```python
import matplotlib.pyplot as plt
from pathlib import Path
import os
import pickle
import numpy as np

cwd = Path.cwd()
path_data = os.path.join(cwd, '../results/')
```

```python
model = 'M0'
data = pickle.load(open(os.path.join(path_data, f'choiceProbab{model}.pkl'), 'rb'))
subjects = list(data.columns)
nsubjects = len(subjects)
cp = np.full((nsubjects, 10), np.nan)
for i, s in enumerate(subjects):
    cp[i] = data[s].values[:10]
plt.plot(range(1, 11), cp.mean(axis=0))
plt.xticks(range(1, 11))
plt.xlabel('trials')
plt.ylabel('choice probability')
plt.show()
```
